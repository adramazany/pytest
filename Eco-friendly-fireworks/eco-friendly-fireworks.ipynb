```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor

# Create presentation
prs = Presentation()

# Define colors
dark_blue = RGBColor(30, 42, 56)
light_gray = RGBColor(229, 229, 229)
green = RGBColor(76, 175, 80)

# Helper function for adding slides
def add_slide(title, content_lines, bg_color=light_gray, title_color=dark_blue, content_color=dark_blue):
    slide_layout = prs.slide_layouts[5]  # blank layout
    slide = prs.slides.add_slide(slide_layout)
    background = slide.background
    fill = background.fill
    fill.solid()
    fill.fore_color.rgb = bg_color

    # Title
    title_box = slide.shapes.add_textbox(Inches(0.8), Inches(0.5), Inches(8.0), Inches(1.0))
    title_tf = title_box.text_frame
    title_tf.text = title
    title_tf.paragraphs[0].font.bold = True
    title_tf.paragraphs[0].font.size = Pt(40)
    title_tf.paragraphs[0].font.color.rgb = title_color

    # Content
    content_box = slide.shapes.add_textbox(Inches(1.0), Inches(1.8), Inches(8.0), Inches(5.0))
    content_tf = content_box.text_frame
    for line in content_lines:
        p = content_tf.add_paragraph()
        p.text = line
        p.font.size = Pt(24)
        p.font.color.rgb = content_color
        p.space_after = Pt(10)

# Slides
slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(slide_layout)
slide.shapes.title.text = "ECO-Friendly Fireworks: The Future of Aerial Shows"
slide.placeholders[1].text = "Sustainable â€¢ Profitable â€¢ Technologically Advanced\n\nAdel Ramezani | Automation\nMÃ¼hlbauer GmbH, Germany"

add_slide("The Problem with Traditional Fireworks", [
    "ğŸŒ«ï¸ Pollution: Air and noise emissions",
    "ğŸ”¥ Safety Risks: Fire hazards and injuries",
    "ğŸ’¸ Single Use: High cost, no reuse",
    "ğŸ§© Sustainability Pressure: Cities seeking eco alternatives"
])

add_slide("Our Solution: Drone-Based Aerial Animation Shows", [
    "â™»ï¸ Eco-friendly and reusable",
    "ğŸ’¡ Programmable animations synchronized with music",
    "ğŸ’» Controlled via intelligent choreography software",
    "ğŸ‡ Visually stunning, customizable for any event"
])

add_slide("Market Opportunity", [
    "ğŸ‡©ğŸ‡ª Over 2,000 German cities host annual fireworks",
    "ğŸ¡ 160â€“180 national & international trade fairs yearly",
    "ğŸ’¶ European drone show market exceeds â‚¬100M",
    "ğŸ“ˆ Strong export potential with 'Made in Germany' quality"
])

add_slide("Why Our Company Is the Right Starting Point", [
    "âš™ï¸ Expertise in industrial automation and robotics",
    "ğŸ§  Skilled engineering & software teams",
    "ğŸ­ Existing manufacturing infrastructure",
    "ğŸ¤ Strong supplier and distribution networks"
])

add_slide("Project Structure", [
    "ğŸ“ˆ Marketing & Business Development",
    "ğŸ’» Drone Choreographing Software",
    "ğŸš Drone Production Line",
    "All three divisions operate synchronously and support each other."
])

add_slide("Implementation Roadmap", [
    "Phase 1 (0â€“6 months): Pilot project with 100 drones",
    "Phase 2 (6â€“18 months): Develop German-made drone and software",
    "Phase 3 (18â€“36 months): Scale production and go international"
])

add_slide("Financial Outlook & ROI", [
    "ğŸ’¶ Initial Investment: â‚¬500kâ€“â‚¬1M",
    "ğŸ“Š Revenue Streams: Shows, Software Licensing, Drone Sales",
    "ğŸ“ˆ Payback Period: 2â€“3 years",
    "ğŸ” Long-term profitability via reusability & service model"
])

add_slide("Future Expansion Potential", [
    "ğŸ¥ Cinematography drones",
    "ğŸš¨ Public safety and rescue drones",
    "ğŸŒ¾ Agricultural monitoring drones",
    "ğŸ“¦ Logistics & transport drones",
    "ğŸ§’ Educational drones for kids",
    "ğŸ§© Sell components and accessories"
])

add_slide("Environmental & Brand Impact", [
    "ğŸŒ± Zero pollution, zero waste",
    "ğŸ”‹ Fully reusable systems",
    "ğŸ‡©ğŸ‡ª German engineering excellence",
    "ğŸŒ Strong sustainability and CSR positioning"
])

add_slide("Next Steps", [
    "âœ… Approve pilot feasibility study",
    "ğŸ¤ Form cross-department task force",
    "ğŸ“ Partner with one municipality",
    "ğŸ“… Define milestones and KPIs"
])

add_slide("Conclusion", [
    "âœ¨ Reinventing the Sky â€” Sustainably âœ¨",
    "Combining German engineering with innovation for eco-friendly entertainment.",
    "Let's make the future of fireworks â€” Made in Germany."
], bg_color=dark_blue, title_color=green, content_color=light_gray)

# Save the presentation
prs.save("ECO_Friendly_Fireworks_Presentation.pptx")
print("âœ… Presentation generated: ECO_Friendly_Fireworks_Presentation.pptx")
```
