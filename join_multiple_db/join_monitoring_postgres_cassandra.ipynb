# cassandra connection

```python
cluster = Cluster(['127.0.0.1'])
session = cluster.connect('monitoring_stream')
```

# postgres connection

```python
engine = create_engine('postgresql://postgres:admin@localhost:5432/muehlbauer.factoryleader')
```

# monitoring_stream.monitoring_data_checkpoint

```python
monitoring_data_checkpoint = pd.DataFrame(list(session.execute('SELECT * FROM monitoring_stream.monitoring_data_checkpoint')))
monitoring_data_checkpoint
```

# monitoring_stream.monitoring_meta_data

```python
monitoring_meta_data = pd.read_sql("select * from monitoring_stream.monitoring_meta_data", engine)
monitoring_meta_data
```

# monitoring_stream.monitoring

```python
monitoring = pd.read_sql("select * from monitoring_stream.monitoring", engine)
monitoring
```

# join monitoring_data_checkpoint and monitoring_meta_data

```python
df10 = monitoring_data_checkpoint.merge(monitoring_meta_data, left_on='monitoring_meta_data_id',right_on='id',how='left')
df10
```

```python
df10[df10.id.isnull()]
```

# join monitoring_data_checkpoint and monitoring

```python
df11 = monitoring_data_checkpoint.merge(monitoring, left_on='monitoring_data_id',right_on='id',how='left')
df11
```
