```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14, 8))
ax.axis('off')

# Define sections of the Business Model Canvas
sections = {
    "Key Partners": " Cloud providers (Google, Dropbox, Nextcloud)\n OCR/AI providers (Google Vision, AWS Textract)\n Accounting software (DATEV, LexOffice)\n GDPR/legal advisors",
    "Key Activities": " AI training & OCR optimization\n Cloud integration & sync\n GDPR compliance & security\n Marketing & partnerships",
    "Value Proposition": "Scan once, find instantly.\n Automatic tagging & categorisation\n Smart folder routing\n Cloud sync (Drive, OneDrive, etc.)\n Instant search & privacy-first",
    "Customer Relationships": " Freemium model (trial + upgrade)\n In-app support & tutorials\n Community feedback portal",
    "Customer Segments": " Freelancers & SMEs\n Families / Individuals\n Accountants & lawyers\n Enterprises needing document automation",
    "Key Resources": " AI/ML engine for document classification\n OCR engine (on-device or cloud)\n Development & design team\n Cloud infrastructure",
    "Channels": " iOS & Android app\n Web dashboard\n Cloud service integrations\n App Store SEO & partnerships",
    "Cost Structure": " Development & AI model training\n Cloud hosting & APIs\n Marketing & customer support\n Legal/GDPR compliance",
    "Revenue Streams": " Subscriptions (4.99/mo  39.99/yr)\n Business plans (1530/user/mo)\n White-label API licensing\n One-time local-only premium purchase"
}

# Define layout positions (x, y, width, height)
layout = {
    "Key Partners": (0.02, 0.55, 0.18, 0.4),
    "Key Activities": (0.22, 0.55, 0.18, 0.4),
    "Value Proposition": (0.42, 0.55, 0.18, 0.4),
    "Customer Relationships": (0.62, 0.55, 0.18, 0.4),
    "Customer Segments": (0.82, 0.55, 0.16, 0.4),
    "Key Resources": (0.02, 0.1, 0.18, 0.35),
    "Channels": (0.22, 0.1, 0.18, 0.35),
    "Cost Structure": (0.42, 0.1, 0.28, 0.35),
    "Revenue Streams": (0.72, 0.1, 0.26, 0.35)
}

# Colors for visual distinction
colors = {
    "Key Partners": "#b3d9ff",
    "Key Activities": "#b3ffb3",
    "Value Proposition": "#fff2b3",
    "Customer Relationships": "#ffb3b3",
    "Customer Segments": "#d9b3ff",
    "Key Resources": "#b3e6ff",
    "Channels": "#ffd9b3",
    "Cost Structure": "#cccccc",
    "Revenue Streams": "#a3a3a3"
}

# Draw boxes and text
for section, (x, y, w, h) in layout.items():
    ax.add_patch(Rectangle((x, y), w, h, linewidth=1.5, edgecolor='black', facecolor=colors[section]))
    ax.text(x + w/2, y + h - 0.02, section, ha='center', va='top', fontsize=13, fontweight='bold')
    ax.text(x + 0.01, y + h - 0.07, sections[section], ha='left', va='top', fontsize=10, wrap=True)

plt.suptitle(" Business Model Canvas  SmartScan App", fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```
