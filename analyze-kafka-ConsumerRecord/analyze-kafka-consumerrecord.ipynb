```python
# splitting string by comma(,) without touching the qouted parts
pieces = [p for p in re.split("(,|\\\".*?\\\"|'.*?')", txt2.strip()[15:-1])]
pieces
```

```python
pieces2 = list(filter(lambda item: item if item!=',' else None , pieces))
pieces2
```

```python
r_pieces2 = list(reversed(pieces2))
pieces3 = []
hold=""
for i, item in enumerate(r_pieces2):
    if isinstance(item, (bytes, bytearray)) :
        item = item
    
    if i<len(r_pieces2)-1 and (r_pieces2[i+1][-1] == "=" or r_pieces2[i+1][-2:] == "=b") :
        hold=item
    else:
        pieces3.insert(0,item+hold)
        hold=""
pieces3
```
